# **Background information:** 

Welcome to the coding challenging! This will be a 4-part challenge and you will develop an text processing techniques that analyze debiasing method that preserves non-discriminative gender-related information. The challenege will be judged based on these three results.

**(a)**Find a dataset containing gender data. 


**(b)**Make a hypothesis/use case for potential gender bias, i.e. demonstrate an example of a discriminatory prediction.

**(c)**Propose a de-biasing method, e.g. by pre-processing data or learning a bias-free model

## **Bonus**
**(d)** Serve the model via an API (bonus).

However, the challenege is not limited to gender bias only. It could also be age discrimination or race discrimination. We are more interested in NLP implementation.





If you have any questions, feel free to reach out to **@Tannistha** :)
